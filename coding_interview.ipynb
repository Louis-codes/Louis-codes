```python
def finding_primes(n):
    primes = []
    a, b = 0, 1

    while len(primes) < n:
        a, b = b, a + b

        if b < 2:
            continue

        prime = True
        for i in range(2, int(b**0.5) + 1):
            if b % i == 0:
                prime = False
                break

        if prime:
            primes.append(b)

    return primes
```

```python
primes = finding_primes(10)
for prime in primes:
    print(prime)
```
